```python
import pandas as pd
```

## doc1

```python
%%time

doc1 = pd.read_csv('doc1.csv')
#doc1.head()

doc1.drop(columns=['Unnamed: 0', 'content', 'title'], inplace=True)
#doc1.head()

#doc1.to_csv('doc1.csv')

indexMap1 = {}
for index, row in doc1.iterrows():
    for token in row['processed_content'].strip("[]").split(', '):
        newKey = str(str(row['id']) + ' ' + token.strip("'"))
        if newKey in indexMap1:
            indexMap1[newKey] = indexMap1[newKey] + 1
        else:
            indexMap1[newKey] = 1
            
index1 = pd.DataFrame.from_dict(indexMap1, orient="index")
index1.reset_index(inplace=True)
new1 = index1["index"].str.split(" ", n = 1, expand = True) 
index1['doc'] = new1[[0]]
index1['token'] = new1[[1]]
index1.drop(columns=['index'], inplace=True)
index1.columns = ['count', 'doc', 'token']
index1 = index1[['doc', 'token', 'count']]
index1.head(20)

#index.head()
index1.to_csv('doc1_index.csv')
```

```python
index1.head(10)
```
